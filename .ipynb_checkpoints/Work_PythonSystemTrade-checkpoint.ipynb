```python
import pandas as pd
import pandas_datareader.data as pdr
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
# Mutral Parameters
start="1949/5/16"
end="2017/1/31"
```

```python
# FredよりNIKKEI平均のデータを取得
N225 = pdr.DataReader("NIKKEI225", 'fred', start, end)
N225.head(1)
```

```python
N225.tail(1)
```

```python
N225.plot(color='darkblue')
plt.ylabel('N225 index')
```

```python
# YahooよりNIKKEI平均を取得
price = pdr.DataReader("^N225", 'yahoo', "1984/1/4", end)
price.head(1)
```

```python
# Fredよりドル円為替レートを取得し、NIKKEI平均との相関をプロット
fx = pdr.DataReader('DEXJPUS','fred',start,end)
port = pd.concat([price.Close, fx], axis=1).dropna()
n = port.Close.pct_change().dropna()
f = port.DEXJPUS.pct_change().dropna()
f.rolling(window=20).corr(n).plot(color="yellow")
plt.ylabel('correlation')
```

```python
#　バーブル以降で区切るNIKKEI225
price1 = price.ix["1990/1/1":]
price1.Close.plot(color='green')
price2 = price.ix["2015",0:2]
price2.tail(1)
plt.ylabel("N225 index")
```

```python
# 日経平均を月単位にResampling
print(price.resample('M').first().tail())
# →月初
```

```python
print(price.resample('M').last().tail())
# →月末
```

```python
print(price.resample('M', loffset='1d').last().tail())
# →データはそのままに日付を１日先に設定する
```

```python
# 年足のデータに変換してチャートを描く
price.resample('A').mean().Close.plot(color='magenta')
plt.ylabel('N225 index')
```

```python
'''
Pandasの記述統計にかんするメソッド
count(計数), quantitle(4分位), sum(合計), mean(平均), median, var(分散), 
std(標準偏差), skew(歪度), kurt(尖度), cumsum(累積合計), diff(差分), pct_change(変化率)
'''
```

```python
# 日経平均の価格変動性
dp = np.log(price.Close).diff()
# →対数変化率　or 対数収益率
vol = dp.std()*np.sqrt(250)
# →1年の対数変化率の標準変化 or ボラティリティ
print(vol, len(price))
```

```python
# Series.rollingを用いて日経平均の終値の250日移動平均を取得する
ma = pd.Series.rolling(price.Close, window=250).mean()
# .max()移動最大値、.min()移動最小値、.median()移動中央値、.std()移動標準偏差
price.Close.plot(label="n225 Close", style='--', color="lightblue")
ma.plot(label='250days ma')
plt.ylabel('N225 index')
plt.legend()
```

```python
# 移動標準偏差を用いる250日文のデータでの移動ボラティリティ
(pd.Series.rolling(np.log(price.Close).diff().dropna(),window=250).std()*np.sqrt(250)).plot()
plt.ylabel('standrd deviation 250 days N225')
# →ボラティリティが上下動きを繰り返し、時として大きくジュアンプしてしばらくそこに留まり、時間が経過すると以前の状態に戻る。
# この現象をボラティリティのクラスタリングと呼ぶ
```
